### Converted timestamps numerical format to datetime string format from both tables to left merge them around Timestamp.

```python
fgi['timestamp']=pd.to_datetime(fgi['timestamp'],unit='s').dt.strftime("%Y-%m-%d")
```

```python
fgi=fgi.rename(columns={'timestamp':'Timestamp','value':'Fear-Greed Index','classification':'Sentiment'})
```

```python
hd['Timestamp']=pd.to_datetime(hd['Timestamp'],unit='ms').dt.strftime("%Y-%m-%d")
```

```python
data = hd[['Timestamp', 'Account', 'Coin', 'Execution Price', 'Size Tokens', 'Size USD', 'Side',
        'Start Position', 'Direction', 'Closed PnL',
     'Crossed', 'Fee'
       ]].merge(fgi[['Timestamp','Fear-Greed Index','Sentiment']],
                      on='Timestamp', how='left')
```

```python
data
```

### Unlabel data where greed-fear index is not known


```python
unlabel_data=data[data.isna().any(axis=1)]
unlabel_data
```

### Calculation of Daily PnL around traders and respective dates.

```python
daily_pnl = data.groupby(['Account','Timestamp'])[['Closed PnL','Fee']].sum().sort_index(level=['Account','Timestamp'],ascending=[True,True])
```

### Calculation of Net PnL which is Daily PnL - Trading Fees

```python
daily_pnl['NetPnL'] = daily_pnl['Closed PnL'] - daily_pnl['Fee']
daily_pnl
```

### Overall Win rate. It is less than 60% Means Therefore some changes in trading strategies are required

```python
data['Won Trade'] = data['Closed PnL'] > 0
```

```python
win_rate = (len(data[data['Won Trade']==True])/len(data))*100
win_rate
```

### Average Trade Size

```python
avg_trade_size=data['Size USD'].mean()
avg_trade_size
```

### There appears to be more trading on Fear day.

```python
trades_per_day=data.groupby('Timestamp')['Account'].count()
trades_per_day1=data.groupby('Timestamp')['Sentiment'].max()
pd.concat(objs=[trades_per_day,trades_per_day1],axis=1)
```

### Calculation of Net Profit when sentiments are extreme, mean profit is less. More risk = more loss. Neutral also have less profit because traders don't have any idea where market will go. Fear and Greed shows better profit as expected because of calculated risk and scope of predictability.

```python
data['NetPnL']=data['Closed PnL']-data['Fee']
```

```python
data.groupby('Sentiment')['NetPnL'].mean()
```

### Less predictabily: Less chances of winning. More predictabily: More chances of winning

```python
data['win']=data['NetPnL']>0
data.groupby('Sentiment')['win'].mean()*100
```

### Box plot of consistent and inconsistent traders PnL distribution. Inconsistent Traders seems to have better profit but extreme volatility. Consistent traders are well calculated, don't take much risk and even though mean profit is lesser than inconsistent traders, It's not much subject to change and profit is positive therefore my suggestion for inconsistent traders is to try to be consistent. 

```python
account_win_rate = data.groupby('Account')['win'].mean()
merged = data.merge(account_win_rate.rename('account_win_rate'), on='Account')
merged['segment_consistency'] = np.where(merged['account_win_rate']>=0.6,'Consistent','Inconsistent')
```

```python
sns.boxplot(x='segment_consistency', y='Closed PnL', data=merged)
plt.title("PnL Distribution: Consistent vs Inconsistent Traders")
plt.show()

merged.groupby('segment_consistency')['Closed PnL'].mean()
```
